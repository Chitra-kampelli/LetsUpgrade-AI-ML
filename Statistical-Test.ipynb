# Statistical Tests 


```python
import pandas as pd
```

```python
dataset =pd.read_csv('general_data.csv')
dataset
```

```python
yes_att = dataset.groupby(dataset.Attrition)
yes = yes_att.get_group("Yes")
print(yes)
```

```python
no_att = dataset.groupby(dataset.Attrition)
No = no_att.get_group("No")
print(No)
```

```python
Attrition_No = pd.DataFrame(No) 
Attrition_No
```

```python
Attrition_Yes = pd.DataFrame(yes)
Attrition_Yes
```

```python
dataset.columns
```

# Attrition Vs Distance from Home

```python
from scipy.stats import mannwhitneyu
```

```python
stat, p=mannwhitneyu(Attrition_Yes.DistanceFromHome ,Attrition_No.DistanceFromHome)
print(stat, p)
```

# Attrition Vs Income

```python
stat, p=mannwhitneyu(Attrition_Yes.MonthlyIncome ,Attrition_No.MonthlyIncome)
print(stat, p)
```

# Attrition Vs Total Working Years

```python

stat, p=mannwhitneyu(Attrition_Yes.TotalWorkingYears ,Attrition_No.TotalWorkingYears)
print(stat, p)
```

 # Attrition Vs Years at company

```python
stat, p=mannwhitneyu(Attrition_Yes.YearsAtCompany ,Attrition_No.YearsAtCompany)
print(stat, p)
```

# Attrition Vs YearsWithCurrentManager

```python
stat, p=mannwhitneyu(Attrition_Yes.YearsWithCurrManager ,Attrition_No.YearsWithCurrManager)
print(stat, p)
```
